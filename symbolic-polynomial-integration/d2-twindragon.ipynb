# Twindragon

```python
import numpy as np
import sympy as sy

import src
```

```python
def twindragon() -> src.SelfAffineSet:
    """Return the Twindragon."""

    r = 1 / sy.sqrt(2)
    T = src.rotation_matrix_2d(sy.pi / 4)

    ifs = [src.contractive_similarity(r, T, sy.Matrix([[x], [0]])) for x in [-1, 1]]

    return src.SelfAffineSet(ifs, src.uniform_measure(len(ifs)), 2)
```

```python
attractor = twindragon()
```

```python
src.render_pre_attractor(attractor, np.array([[-1, 0, 1, 0], [0, -1, 0, 1]]), 3)
```
