# Cube

```python
from itertools import product

import sympy as sy

import src
```

```python
def cube(a: sy.Expr, b: sy.Expr) -> src.SelfAffineSet:
    """Return the cube [a, b]x[a, b]x[a, b]."""

    ifs = [
        src.contractive_similarity(sy.Rational(1, 2), sy.eye(3), sy.Matrix(p))
        for p in product((a, b), (a, b), (a, b))
    ]

    return src.SelfAffineSet(ifs, src.uniform_measure(len(ifs)), 3)
```

```python
attractor = cube(0, 1)
```

```python
src.render_chaos_game(attractor, 10_000)
```

```python
tot_deg_max = 4

values = src.compute_polynomial_integral(attractor, tot_deg_max)

src.display_values(
    values, attractor.space_dim, variables=sy.symbols("x y z", real=True)
)
```
