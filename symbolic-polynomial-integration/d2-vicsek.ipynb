# Vicsek fractal 2d

```python
from itertools import product

import numpy as np
import sympy as sy

import src
```

```python
def vicsek_2d(factor: sy.Expr, inner_rotation: bool = False) -> src.SelfAffineSet:
    """Return the Vicsek 2d."""

    R = src.rotation_matrix_2d(sy.pi / 4) if inner_rotation else sy.eye(2)
    S0 = src.contractive_similarity(factor, R, sy.Matrix([[0], [0]]))

    ifs = [
        S0,
        *[
            src.contractive_similarity(factor, sy.eye(2), sy.Matrix(p))
            for p in product((-1, 1), (-1, 1))
        ],
    ]

    return src.SelfAffineSet(ifs, src.uniform_measure(len(ifs)), 2)
```

```python
attractor = vicsek_2d(sy.Rational(1, 3))
```

```python
src.render_pre_attractor(attractor, np.array([[-1, 1, 1, -1], [-1, -1, 1, 1]]), 3)
```

```python
src.render_chaos_game(attractor, 10_000)
```

```python
tot_deg_max = 4

values = src.compute_polynomial_integral(attractor, tot_deg_max)

src.display_values(values, attractor.space_dim, variables=sy.symbols("x y", real=True))
```
