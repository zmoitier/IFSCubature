# Durer's Pentagons

```python
import numpy as np
import sympy as sy

import src
```

```python
def durer_pentagons() -> src.SelfAffineSet:
    """Return the Durer's Pentagons."""

    r = (3 - sy.sqrt(5)) / 2
    ifs = [
        src.contractive_similarity(r, -sy.eye(2), sy.zeros(2, 1)),
        *[
            src.contractive_similarity(
                r,
                sy.eye(2),
                sy.Matrix(
                    [
                        [sy.cos(2 * i * sy.pi / 5).trigsimp()],
                        [sy.sin(2 * i * sy.pi / 5).trigsimp()],
                    ]
                ),
            )
            for i in range(5)
        ],
    ]

    return src.SelfAffineSet(ifs, src.uniform_measure(len(ifs)), 2)
```

```python
attractor = durer_pentagons()
```

```python
src.render_pre_attractor(
    attractor,
    np.array(
        [
            [np.cos(2 * i * np.pi / 5) for i in range(5)],
            [np.sin(2 * i * np.pi / 5) for i in range(5)],
        ]
    ),
    3,
)
```
