# Vicsek fractal 3d

```python
from itertools import product

import sympy as sy

import src
```

```python
def vicsek_3d(factor: sy.Expr, inner_rotation: bool = False) -> src.SelfAffineSet:
    """Return the Vicsek 3d."""

    if inner_rotation:
        R = src.rotation_matrix_3d(-sy.pi / 4, (0, 0, 1)) * src.rotation_matrix_3d(
            sy.atan(sy.sqrt(2)), (0, 1, 0)
        )
    else:
        R = sy.eye(3)

    S0 = src.contractive_similarity(factor, R, sy.Matrix([[0], [0], [0]]))

    ifs = [
        S0,
        *[
            src.contractive_similarity(factor, sy.eye(3), sy.Matrix(p))
            for p in product((-1, 1), (-1, 1), (-1, 1))
        ],
    ]

    return src.SelfAffineSet(ifs, src.uniform_measure(len(ifs)), 3)
```

```python
attractor = vicsek_3d(sy.Rational(1, 3))
```

```python
src.render_chaos_game(attractor, 10_000)
```

```python
tot_deg_max = 4

values = src.compute_polynomial_integral(attractor, tot_deg_max)

src.display_values(
    values, attractor.space_dim, variables=sy.symbols("x y z", real=True)
)
```
