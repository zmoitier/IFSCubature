# Terdragon

```python
import numpy as np
import sympy as sy

import src
```

```python
def terdragon() -> src.SelfAffineSet:
    """Return the Terdragon."""

    r = 1 / sy.sqrt(3)
    Ts = [
        src.rotation_matrix_2d(sy.pi / 6),
        src.rotation_matrix_2d(-sy.pi / 2),
        src.rotation_matrix_2d(sy.pi / 6),
    ]
    xs = [-1, 0, 1]

    ifs = [
        src.contractive_similarity(r, T, sy.Matrix([[x], [0]]))
        for T, x in zip(Ts, xs, strict=True)
    ]

    return src.SelfAffineSet(ifs, src.uniform_measure(len(ifs)), 2)
```

```python
attractor = terdragon()
```

```python
y = 1 / np.sqrt(3)
src.render_pre_attractor(attractor, np.array([[-1, 0, 1, 0], [0, -y, 0, y]]), 3)
```
