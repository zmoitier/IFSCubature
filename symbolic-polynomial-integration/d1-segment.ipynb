# Segment

```python
import numpy as np
import sympy as sy

import src
```

```python
def segment(a: sy.Expr, b: sy.Expr) -> src.SelfAffineSet:
    """Return the segment [a, b]."""

    ifs = tuple(
        src.contractive_similarity(sy.Rational(1, 2), sy.eye(1), sy.Matrix([c]))
        for c in [a, b]
    )

    return src.SelfAffineSet(ifs, src.uniform_measure(len(ifs)), 1)
```

```python
attractor = segment(0, 1)
```
