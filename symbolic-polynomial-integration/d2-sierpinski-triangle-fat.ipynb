# Fat SierpiÅ„ski triangle

```python
import numpy as np
import sympy as sy

import src
```

```python
def sierpinski_triangle(factor: sy.Expr | None = None) -> src.SelfAffineSet:
    """Return the Sierpinski triangle."""

    if factor is None:
        factor = sy.Rational(1, 2)

    ifs = [
        src.contractive_similarity(factor, sy.eye(2), sy.Matrix(p))
        for p in [
            (1, 0),
            (sy.Rational(-1, 2), sy.sqrt(3) / 2),
            (sy.Rational(-1, 2), -sy.sqrt(3) / 2),
        ]
    ]

    return src.SelfAffineSet(ifs, src.uniform_measure(len(ifs)), 2)
```

```python
attractor = sierpinski_triangle((sy.sqrt(5) - 1) / 2)

# a = (17 + 3 * sy.sqrt(33)) ** sy.Rational(1, 3)
# attractor = sierpinski_triangle(a / 3 - 2 / (3 * a) - sy.Rational(1, 3))
```
