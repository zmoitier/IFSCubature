# Brick

```python
import numpy as np
import sympy as sy

import src
```

```python
def brick() -> src.SelfAffineSet:
    """Return the brick."""

    r = sy.Rational(1, 3)

    ifs = [
        src.contractive_similarity(r, sy.eye(2), sy.Matrix(p))
        for p in [
            (-1, -1),
            (1, -1),
            (0, 0),
            (1, 0),
            (2, 0),
            (-1, 1),
            (0, 1),
            (1, 1),
            (0, 2),
        ]
    ]

    return src.SelfAffineSet(ifs, src.uniform_measure(len(ifs)), 2)
```

```python
attractor = brick()
```

```python
src.render_pre_attractor(attractor, np.array([[-1, 1, 1, -1], [-1, -1, 1, 1]]), 3)
```
