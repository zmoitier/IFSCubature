# Heighway Dragon

```python
import numpy as np
import sympy as sy

import src
```

```python
def heighway_dragon() -> src.SelfAffineSet:
    """Return the Heighway Dragon."""

    r = 1 / sy.sqrt(2)
    ifs = [
        src.contractive_similarity(
            r, src.rotation_matrix_2d(sy.pi / 4), sy.Matrix([[0], [0]])
        ),
        src.contractive_similarity(
            r,
            src.rotation_matrix_2d(3 * sy.pi / 4),
            sy.Matrix([[sy.Rational(3, 5)], [sy.Rational(1, 5)]]),
        ),
    ]

    return src.SelfAffineSet(ifs, src.uniform_measure(len(ifs)), 2)
```

```python
attractor = heighway_dragon()
```
