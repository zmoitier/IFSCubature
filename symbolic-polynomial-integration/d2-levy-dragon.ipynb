# Lévy Dragon

```python
import numpy as np
import sympy as sy

import src
```

```python
def levy_dragon() -> src.SelfAffineSet:
    """Return the Lévy Dragon."""

    r = 1 / sy.sqrt(2)
    ifs = [
        src.contractive_similarity(
            r, src.rotation_matrix_2d(sy.pi / 4), sy.Matrix([[-1], [0]])
        ),
        src.contractive_similarity(
            r, src.rotation_matrix_2d(-sy.pi / 4), sy.Matrix([[1], [0]])
        ),
    ]

    return src.SelfAffineSet(ifs, src.uniform_measure(len(ifs)), 2)
```

```python
attractor = levy_dragon()
```
