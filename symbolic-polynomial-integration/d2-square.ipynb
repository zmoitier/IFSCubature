# Square

```python
from itertools import product

import numpy as np
import sympy as sy

import src
```

```python
def square(a: sy.Expr, b: sy.Expr) -> src.SelfAffineSet:
    """Return the square [a, b]x[a, b]."""

    ifs = [
        src.contractive_similarity(sy.Rational(1, 2), sy.eye(2), sy.Matrix(p))
        for p in product((a, b), (a, b))
    ]

    return src.SelfAffineSet(ifs, src.uniform_measure(len(ifs)), 2)
```

```python
attractor = square(-1, 1)
```

```python
src.render_pre_attractor(attractor, np.array([[-1, 1, 1, -1], [-1, -1, 1, 1]]), 1)
```

```python
src.render_chaos_game(attractor, 10_000)
```
