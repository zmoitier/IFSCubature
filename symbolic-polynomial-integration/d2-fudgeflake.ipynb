# Fudgeflake

```python
import numpy as np
import sympy as sy

import src
```

```python
def fudgeflake() -> src.SelfAffineSet:
    """Return the Fudgeflake."""

    r = 1 / sy.sqrt(3)
    T = src.rotation_matrix_2d(sy.pi / 6)

    ifs = [
        src.contractive_similarity(r, T, sy.Matrix(p))
        for p in [
            (1, 0),
            (sy.Rational(-1, 2), sy.sqrt(3) / 2),
            (sy.Rational(-1, 2), -sy.sqrt(3) / 2),
        ]
    ]

    return src.SelfAffineSet(ifs, src.uniform_measure(len(ifs)), 2)
```

```python
attractor = fudgeflake()
```

```python
src.render_pre_attractor(
    attractor,
    np.array(
        [
            [np.cos(t * np.pi / 3) for t in range(6)],
            [np.sin(t * np.pi / 3) for t in range(6)],
        ]
    ),
    3,
)
```
