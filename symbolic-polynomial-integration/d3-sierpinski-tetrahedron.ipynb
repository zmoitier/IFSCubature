# SierpiÅ„ski tetrahedron

```python
import sympy as sy

import src
```

```python
def sierpinski_tetrahedron(factor: sy.Expr | None = None) -> src.SelfAffineSet:
    """Return the Sierpinski tetrahedron."""

    if factor is None:
        factor = sy.Rational(1, 2)

    ifs = [
        src.contractive_similarity(factor, sy.eye(3), sy.Matrix(p))
        for p in [
            (1, 0, -sy.sqrt(2) / 2),
            (-1, 0, -sy.sqrt(2) / 2),
            (0, 1, sy.sqrt(2) / 2),
            (0, -1, sy.sqrt(2) / 2),
        ]
    ]

    return src.SelfAffineSet(ifs, src.uniform_measure(len(ifs)), 3)
```

```python
attractor = sierpinski_tetrahedron(sy.Rational(1, 2))
```
