# Triangle

```python
import numpy as np
import sympy as sy

import src
```

```python
def triangle() -> src.SelfAffineSet:
    """Return the triangle."""

    ifs = [
        src.contractive_similarity(sy.Rational(1, 2), -sy.eye(2), sy.Matrix((0, 0))),
        *[
            src.contractive_similarity(sy.Rational(1, 2), sy.eye(2), sy.Matrix(c))
            for c in [
                (1, 0),
                (sy.Rational(-1, 2), sy.sqrt(3) / 2),
                (sy.Rational(-1, 2), -sy.sqrt(3) / 2),
            ]
        ],
    ]

    return src.SelfAffineSet(ifs, src.uniform_measure(len(ifs)), 2)
```

```python
attractor = triangle()
```

```python
src.render_pre_attractor(
    attractor, np.array([[1, -0.5, -0.5], [0, np.sqrt(3) / 2, -np.sqrt(3) / 2]]), 1
)
```

```python
src.render_chaos_game(attractor, 10_000)
```
