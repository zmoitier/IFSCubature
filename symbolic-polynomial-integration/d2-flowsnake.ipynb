# Gosper's flowsnake

```python
import numpy as np
import sympy as sy

import src
```

```python
def flowsnake() -> src.SelfAffineSet:
    """Return the Gosper's flowsnake."""

    factor = 1 / sy.sqrt(7)
    R = src.rotation_matrix_2d(sy.atan(sy.sqrt(3) / 5))

    ifs = [
        src.contractive_similarity(factor, R, sy.Matrix([[0], [0]])),
        *[
            src.contractive_similarity(factor, R, sy.Matrix(p))
            for p in [(sy.cos(t * sy.pi / 3), sy.sin(t * sy.pi / 3)) for t in range(6)]
        ],
    ]

    return src.SelfAffineSet(ifs, src.uniform_measure(len(ifs)), 2)
```

```python
attractor = flowsnake()
```

```python
src.render_pre_attractor(
    attractor,
    np.array(
        [
            [np.cos(t * np.pi / 3) for t in range(6)],
            [np.sin(t * np.pi / 3) for t in range(6)],
        ]
    ),
    3,
)
```
